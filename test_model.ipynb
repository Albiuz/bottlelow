```python
def _get_levels(pred):
    # get region
    region_range = 7
    step = 5
    index = 0
    bboxs = []
    for i in range(0,pred.shape[1]-region_range, step):
        score = 0
        for l in range(region_range):
            score += pred[0][i+l][0][0]
        if score > 0.5*region_range:
            index = i+region_range//2
            bboxs.append({'score':score,'index':index})
    return bboxs
```
